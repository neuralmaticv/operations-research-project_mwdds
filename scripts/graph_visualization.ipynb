```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def read_graph(instance_path, num_vertices):
    with open(instance_path, 'r') as f:
        vertices = {}
        counter = 0

        for line in f:
            counter += 1
            vertex_id, vertex_weight = line.strip().split()
            vertices[int(vertex_id)] = int(vertex_weight)
            if counter == num_vertices:
                break

        # Citaj grane
        edges = []
        for line in f:
            fields = line.strip().split()
            source, target = fields
            edges.append((int(source), int(target)))

    return vertices, edges
```

```python
for i in range(0, 10):
    instance_id = i
    vertices, edges = read_graph("../instances/level0/instance" + str(i) + ".txt", 5)
    G = nx.DiGraph(edges)

    for i in vertices:
        G.add_node(i, weight=vertices[i])

    labels = {n: str(n) + '\n w=' + str(G.nodes[n]['weight']) for n in G.nodes}
    nx.draw(G, labels=labels, node_size=2500, font_weight='bold')
    plt.savefig("../instances/level0/imgs/instance" + str(instance_id) + ".png", format="PNG")
    plt.show()
    plt.close()
```
